```python
def classifier_nappe_fluctuation(df, col="niveau_nappe_eau"):

    niveau = df[col].dropna()

    # amplitude totale
    amplitude = niveau.max() - niveau.min()

    # variations mensuelles
    variations = niveau.diff().dropna()

    variation_moy = variations.abs().mean()
    variabilite = variations.std()

    # pente moyenne (tendance globale)
    x = np.arange(len(niveau))
    slope = np.polyfit(x, niveau, 1)[0]

    # indice de dynamique
    indice = variation_moy + variabilite

    # classification
    if indice > 1.0:
        type_nappe = "nappe rÃĐactive"
    elif indice > 0.4:
        type_nappe = "nappe intermÃĐdiaire"
    else:
        type_nappe = "nappe inertielle"

    return {
        "amplitude": amplitude,
        "variation_moyenne": variation_moy,
        "variabilite": variabilite,
        "pente": slope,
        "indice_dynamique": indice,
        "type_nappe": type_nappe
    }
```

## Interpolations

Lâinterpolation est une technique mathÃĐmatique qui permet dâestimer des valeurs manquantes ou intermÃĐdiaires Ã  partir dâun ensemble de points connus. LâidÃĐe est de construire une fonction qui relie ces points, de faÃ§on Ã  pouvoir calculer des valeurs pour nâimporte quel point situÃĐ entre eux.

### Interpolation linÃĐaire

Lâinterpolation linÃĐaire permet dâestimer des valeurs manquantes en supposant que lâÃĐvolution entre deux points connus suit une droite. Les valeurs manquantes sont donc calculÃĐes Ã  partir de la droite reliant les observations voisine

$$
f(x) = ax + b
$$

oÃđ :

- $a$ reprÃĐsente la pente 
- $b$ lâordonnÃĐe Ã  lâorigine

```python
def interpolation_lineaire_array(df:pd.DataFrame, valeur_de_travail:str)->np.ndarray:
    """Interpole linÃĐairement la colonne `valeur_de_travail`

    Args:
        df (pd.DataFrame): Dataset Ã  complÃĐter
        valeur_de_travail (str): Valeur que l'on veut interpolÃĐ

    Returns:
        np.ndarray: Le dataset complÃĐtÃĐ sur la valeur demandÃĐe
    """

    return df[valeur_de_travail].interpolate(method='linear', limit_direction='both').to_numpy()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['time'], interpolation_lineaire_array(df, valeur_de_travail), label="PrÃĐdiction")
plt.plot(df['time'], df[valeur_de_travail], label="Mesure")
plt.xlabel("Date")
plt.ylabel(valeur_de_travail)
plt.legend()
plt.show()
```

### Interpolation via spline cubique

Lâinterpolation cubique estime les valeurs manquantes en utilisant un polynÃīme de degrÃĐ 3 entre les points connus. Contrairement Ã  lâinterpolation linÃĐaire, cette mÃĐthode permet dâobtenir une courbe plus lisse et de mieux capturer les variations non linÃĐaires des donnÃĐes.

$$
f(x) = ax^3 + bx^2 + cx + d
$$

oÃđ :

- $a$, $b$, $c$ et $d$ sont des coefficients dÃĐterminÃĐs Ã  partir des points connus afin d'ajuster le polynÃīme aux donnÃĐes.

```python
def interpolation_cubique_array(df:pd.DataFrame, valeur_de_travail:str)->np.ndarray:
    """Interpole cubiquement la colonne `valeur_de_travail`

    Args:
        df (pd.DataFrame): Dataset Ã  complÃĐter
        valeur_de_travail (str): Valeur que l'on veut interpolÃĐ

    Returns:
        np.ndarray: Le dataset complÃĐtÃĐ sur la valeur demandÃĐe
    """

    return df[valeur_de_travail].interpolate(method='cubic').to_numpy()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['time'], interpolation_cubique_array(df, valeur_de_travail), label="PrÃĐdiction")
plt.plot(df['time'], df[valeur_de_travail], label="Mesure")
plt.xlabel("Date")
plt.ylabel(valeur_de_travail)
plt.legend()
plt.show()
```

### Interpolation via PolynÃīmes de degrÃĐ n

Lâinterpolation polynomiale estime les valeurs manquantes en utilisant un polynÃīme de degrÃĐ $n$ ajustÃĐ aux points connus.

$$
f(x) = a_n x^n + a_{n-1} x^{n-1} + \dots + a_1 x + a_0
$$

oÃđ :

- $n$ est le degrÃĐ du polynÃīme,
- $a_0, a_1, \dots, a_n$ sont des coefficients dÃĐterminÃĐs Ã  partir des points connus.

```python
def interpolation_polynomiale_array(df:pd.DataFrame, valeur_de_travail:str, deg:int=2)->np.ndarray:
    """Interpole polynomialement la colonne `valeur_de_travail`

    Args:
        df (pd.DataFrame): Dataset Ã  complÃĐter
        valeur_de_travail (str): Valeur que l'on veut interpolÃĐ
        deg (int): DegrÃĐ de la polynomiale

    Returns:
        np.ndarray: Le dataset complÃĐtÃĐ sur la valeur demandÃĐe
    """

    return df[valeur_de_travail].interpolate(method="polynomial", order=deg).to_numpy()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['time'], interpolation_polynomiale_array(df, valeur_de_travail), label="PrÃĐdiction")
plt.plot(df['time'], df[valeur_de_travail], label="Mesure")
plt.xlabel("Date")
plt.ylabel(valeur_de_travail)
plt.legend()
plt.show()
```

### Interpolation via B-spline

Lâinterpolation par spline estime les valeurs manquantes en utilisant des polynÃīmes de degrÃĐ $n$ dÃĐfinis sur plusieurs intervalles entre les points connus. Cela permet dâobtenir une courbe plus lisse et plus stable quâune interpolation polynomiale globale.

$$
f(x) = ax^3 + bx^2 + cx + d
$$

oÃđ :

- $a$, $b$, $c$ et $d$ sont des coefficients dÃĐterminÃĐs Ã  partir des points connus.

```python
def interpolation_spline_array(df:pd.DataFrame, valeur_de_travail:str, deg:int=3)->np.ndarray:
    """Interpole via une spline polynomiale par moindres carrÃĐs la colonne `valeur_de_travail`

    Args:
        df (pd.DataFrame): Dataset Ã  complÃĐter
        valeur_de_travail (str): Valeur que l'on veut interpolÃĐ
        deg (int): DegrÃĐ de la spline

    Returns:
        np.ndarray: Le dataset complÃĐtÃĐ sur la valeur demandÃĐe
    """

    return df[valeur_de_travail].interpolate(method="spline", order=deg).to_numpy()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['time'], interpolation_spline_array(df, valeur_de_travail), label="PrÃĐdiction")
plt.plot(df['time'], df[valeur_de_travail], label="Mesure")
plt.xlabel("Date")
plt.ylabel(valeur_de_travail)
plt.legend()
plt.show()
```

### Interpolation via PCHIP

Lâinterpolation PCHIP (Piecewise Cubic Hermite Interpolating Polynomial) utilise des polynÃīmes cubiques **par morceaux** pour estimer les valeurs manquantes, en prÃĐservant la **monotonie** et les **formes locales** de la sÃĐrie.  

$$
f(x) = S_i(x), \quad x \in [x_i, x_{i+1}]
$$

oÃđ :

- $S_i(x)$ est un polynÃīme cubique dÃĐfini sur chaque intervalle $[x_i, x_{i+1}]$.

```python
def interpolation_pchip_array(df:pd.DataFrame, valeur_de_travail:str)->np.ndarray:
    """Interpole via PCHIP la colonne `valeur_de_travail`

    Args:
        df (pd.DataFrame): Dataset Ã  complÃĐter
        valeur_de_travail (str): Valeur que l'on veut interpolÃĐ

    Returns:
        np.ndarray: Le dataset complÃĐtÃĐ sur la valeur demandÃĐe
    """

    return df[valeur_de_travail].interpolate(method='pchip').to_numpy()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['time'], interpolation_pchip_array(df, valeur_de_travail), label="PrÃĐdiction")
plt.plot(df['time'], df[valeur_de_travail], label="Mesure")
plt.xlabel("Date")
plt.ylabel(valeur_de_travail)
plt.legend()
plt.show()
```

### Interpolation via akima

Lâinterpolation Akima utilise ÃĐgalement des polynÃīmes cubiques **par morceaux**, mais avec un calcul spÃĐcial des pentes pour ÃĐviter les oscillations excessives. Elle est utile pour les donnÃĐes prÃĐsentant des **variations abruptes**.  

$$
f(x) = A_i(x), \quad x \in [x_i, x_{i+1}]
$$

oÃđ :

- $A_i(x)$ est le polynÃīme cubique sur lâintervalle $[x_i, x_{i+1}]$ avec des pentes calculÃĐes selon la mÃĐthode dâAkimaÂđ.

mÃĐthode d'AkimaÂđ : Les pentes aux points intermÃĐdiaires ne sont pas simplement calculÃĐes par la pente linÃĐaire des segments voisins, mais selon un moyen pondÃĐrÃĐ des pentes locales des segments adjacents 

```python
def interpolation_akima_array(df:pd.DataFrame, valeur_de_travail:str)->np.ndarray:
    """Interpole via Akima la colonne `valeur_de_travail`

    Args:
        df (pd.DataFrame): Dataset Ã  complÃĐter
        valeur_de_travail (str): Valeur que l'on veut interpolÃĐ

    Returns:
        np.ndarray: Le dataset complÃĐtÃĐ sur la valeur demandÃĐe
    """

    return df[valeur_de_travail].interpolate(method='akima').to_numpy()
```

```python
def charger_et_preparer_donnees(dossier_path, features):
    # 1. Chargement global (comme tu faisais)
    fichiers_csv = glob.glob(os.path.join(dossier_path, "*.csv"))
    liste_df = [pd.read_csv(f, sep=';') for f in fichiers_csv]
    df_final = pd.concat(liste_df, ignore_index=True)
    
    # 2. Nettoyage et Tri
    df_final['time'] = pd.to_datetime(df_final['time'])
    # On trie par station PUIS par temps pour que les chronologies soient respectÃĐes
    df_final = df_final.sort_values(['code_bss', 'time'])
    
    df_travail = df_final.copy()
    # Conversion du temps pour le scaler
    df_travail['time_num'] = df_travail['time'].astype('int64') // 10**9
    
    # On ajuste la liste des features si besoin (on utilise time_num au lieu de time)
    features_scaler = [f if f != 'time' else 'time_num' for f in features]
    
    scaler = MinMaxScaler(feature_range=(-1, 1))
    df_travail[features_scaler] = scaler.fit_transform(df_travail[features_scaler])
    
    return df_travail, scaler

def creer_sequences_par_bss(df, features, window_size):
    X_list, y_list = [], []
    
    # On groupe par code_bss pour traiter chaque piÃĐzomÃĻtre sÃĐparÃĐment
    for _, group in df.groupby('code_bss'):
        data = group[features].values
        
        # On ne crÃĐe des fenÃŠtres que si la station a assez de donnÃĐes
        if len(data) > window_size:
            for i in range(window_size, len(data)):
                X_list.append(data[i-window_size:i, :])
                y_list.append(data[i, :])
    
    return np.array(X_list), np.array(y_list)

# --- EXECUTION ---

# 1. PrÃĐparation
features = ["niveau_nappe_eau","lon","lat","time","ETP_Q","PRELIQ_Q","T_Q","surface_imp","surface_totale"]
df_norm, mon_scaler = charger_et_preparer_donnees(dossier_nappe, features)

# 2. CrÃĐation des fenÃŠtres ÃĐtanches (6 mois ici)
features_pour_ia = [f if f != 'time' else 'time_num' for f in features]
X, y = creer_sequences_par_bss(df_norm, features_pour_ia, window_size=24)

# 3. Gestion des NaN pour la Masked Loss
X = np.nan_to_num(X, nan=-999.0)
y = np.nan_to_num(y, nan=-999.0)

# 4. Split Train/Val
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.5, random_state=42)

@keras.saving.register_keras_serializable()
def masked_mse(y_true, y_pred):
    # On crÃĐe un masque : True lÃ  oÃđ la donnÃĐe n'est pas la valeur sentinelle
    mask = torch.logical_not(torch.eq(y_true, -999.0))
    
    # On ne garde que les valeurs valides
    y_true_masked = torch.masked_select(y_true, mask)
    y_pred_masked = torch.masked_select(y_pred, mask)
    
    # Calcul de l'erreur classique (MSE) sur ce qu'il reste
    return torch.mean(torch.square(y_true_masked - y_pred_masked))
```

```python
def lstm_predict_array(df_entree, model, scaler, features, window_size, target_col="T_Q"):
    df_travail = df_entree.copy()
    
    # 1. PrÃĐparation du temps : On ÃĐcrase 'time' par sa version numÃĐrique
    # pour que le nom de la colonne soit identique Ã  celui vu au 'fit'
    if 'time' in df_travail.columns:
        df_travail['time'] = pd.to_datetime(df_travail['time']).astype('int64') // 10**9
    
    # 2. On s'assure de ne passer QUE les colonnes du scaler, dans le bon ordre
    # Le scaler attend les noms : ["niveau_nappe_eau", ..., "time", ...]
    df_pour_scaler = df_travail[features]
    
    # 3. Transformation (maintenant les noms correspondent)
    values_norm = scaler.transform(df_pour_scaler)
    
    # 4. Gestion des NaN pour le modÃĻle
    values_masked = np.nan_to_num(values_norm, nan=-999.0)
    
    # 5. CrÃĐation des sÃĐquences
    X_all = []
    for i in range(window_size, len(values_masked)):
        X_all.append(values_masked[i-window_size:i, :])
    X_all = np.array(X_all, dtype='float32')
    
    if len(X_all) == 0:
        return np.full(len(df_entree), np.nan)
    
    # 6. PrÃĐdiction
    y_pred_norm = model.predict(X_all, verbose=0)
    
    # 7. Inverse Transform
    y_pred_final = scaler.inverse_transform(y_pred_norm)
    
    # 8. Extraction avec l'index dynamique
    target_idx = features.index(target_col)
    
    # 9. Reconstruction
    full_array = np.full(len(df_entree), np.nan)
    full_array[window_size:] = y_pred_final[:, target_idx]
    
    return full_array
```

### Random forest

La Random Forest construit plusieurs arbres de dÃĐcision sur des ÃĐchantillons diffÃĐrents du jeu de donnÃĐes. Chaque arbre fait une prÃĐdiction indÃĐpendante. Pour la rÃĐgression, la Random Forest prend la moyenne des prÃĐdictions de tous les arbres. Chaque arbre utilise seulement un sous-ensemble alÃĐatoire de variables Ã  chaque dÃĐcision, ce qui rend le modÃĻle plus robuste et rÃĐduit le risque de surapprentissage.

```python
def random_forest_delta_array(df:pd.DataFrame, valeur_de_travail:str, features:list[str]=None)->np.ndarray:
    """
    Interpole une colonne cible dâun DataFrame en utilisant un modÃĻle 
    Random Forest pour prÃĐdire le taux de variation journalier et 
    reconstruire la sÃĐrie temporelle de maniÃĻre itÃĐrative.

    Args:
        df (pd.DataFrame): Dataset Ã  complÃĐter
        valeur_de_travail (str): Valeur que l'on veut interpolÃĐ
        features (list[str], optional): Liste des colonnes Ã  utiliser comme 
            caractÃĐristiques pour entraÃŪner le modÃĻle de prÃĐdiction des deltas.
            Par dÃĐfaut ['ETP_Q', 'T_Q'].

    Returns:
        np.ndarray: Le dataset complÃĐtÃĐ sur la valeur demandÃĐe
    """

    if features is None:
        features = ['ETP_Q', 'T_Q']

    df = df.copy()
    df['time'] = pd.to_datetime(df['time'])
    df = df.sort_values('time').reset_index(drop=True)

    # Calcul prÃĐcis du temps ÃĐcoulÃĐ en jours (mÃŠme si c'est mensuel)
    # total_seconds / 86400 donne le nombre exact de jours entre deux relevÃĐs
    df['days_diff'] = df['time'].diff().dt.total_seconds() / 86400
    df['delta_niveaux'] = df[valeur_de_travail].diff()
    df['delta_per_day'] = df['delta_niveaux'] / df['days_diff']

    # Nettoyage pour l'entraÃŪnement
    # On ne garde que les lignes oÃđ on a Ã  la fois la cible et les features
    train_df = df.dropna(subset=[valeur_de_travail, 'delta_per_day'] + features)

    if len(train_df) < 3:
        # Si trop peu de donnÃĐes, on renvoie des NaN (le script passera au suivant)
        return np.full(len(df), np.nan)

    # ModÃĻle
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(train_df[features], train_df['delta_per_day'])

    # PrÃĐdiction des deltas (on remplace les features manquantes par la moyenne)
    X_all = df[features].fillna(train_df[features].mean())
    pred_delta_per_day = model.predict(X_all)

    # Reconstruction itÃĐrative du signal
    y_reconstruit = df[valeur_de_travail].to_numpy().copy()

    # Si la 1ÃĻre valeur est manquante, on cherche la 1ÃĻre connue pour reculer ou on prend la moyenne
    if np.isnan(y_reconstruit[0]):
        first_valid_idx = df[valeur_de_travail].first_valid_index()
        if first_valid_idx is not None:
            y_reconstruit[0] = df[valeur_de_travail].iloc[first_valid_idx]
        else:
            y_reconstruit[0] = train_df[valeur_de_travail].mean()

    for i in range(1, len(y_reconstruit)):
        if np.isnan(y_reconstruit[i]):
            # On applique : Valeur_prÃĐcÃĐdente + (Pente_prÃĐdite * Nb_jours_rÃĐels)
            d_days = df['days_diff'].iloc[i]
            y_reconstruit[i] = y_reconstruit[i-1] + (pred_delta_per_day[i] * d_days)

    return y_reconstruit
```

### KNN

La mÃĐthode des K plus proches voisins (KNN) permet dâestimer la valeur dâun point en regardant les K points les plus proches dans le jeu de donnÃĐes.
Pour la rÃĐgression, on prend la moyenne des valeurs de ces voisins, et pour la classification, on prend la classe la plus frÃĐquente.
La distance entre les points est souvent calculÃĐe avec la distance Euclidienne.

```python
def knn_impute(df:pd.DataFrame, valeur_de_travail:str, k:int=5, past_only:bool=False)->np.ndarray:
    """
    Impute les valeurs manquantes d'une colonne cible en utilisant la mÃĐthode
    des K plus proches voisins (KNN) avec normalisation des caractÃĐristiques.
    
    La fonction peut restreindre l'imputation aux valeurs passÃĐes uniquement 
    (option `past_only`) pour respecter l'ordre temporel.

    Args:
        df (pd.DataFrame): Dataset Ã  complÃĐter
        valeur_de_travail (str): Valeur que l'on veut interpolÃĐ
        k (int, optional): Nombre de voisins Ã  considÃĐrer 
            Par dÃĐfaut : 5.
        past_only (bool, optional): Si True, seuls les points temporellement 
            antÃĐrieurs Ã  la ligne Ã  remplir sont utilisÃĐs pour l'imputation. 
            Par dÃĐfaut : False.

    Returns:
        np.ndarray: Le dataset complÃĐtÃĐ sur la valeur demandÃĐe
    """
    df = df.copy()
    
    # PrÃĐparation des features
    df["time"] = pd.to_datetime(df["time"])
    df["year"] = df["time"].dt.year
    df["month"] = df["time"].dt.month
    df["month_sin"] = np.sin(2*np.pi*df["month"]/12)
    df["month_cos"] = np.cos(2*np.pi*df["month"]/12)
    
    df = df.sort_values(["code_bss", "time"])
    
    features = ["lon", "lat", "year", "month_sin", "month_cos", 
                ]

    # SÃĐparation et Nettoyage
    # On dÃĐfinit 'complete' : les lignes oÃđ on a la cible ET toutes les features
    # C'est crucial pour le StandardScaler
    mask_complete = df[valeur_de_travail].notna() & df[features].notna().all(axis=1)
    complete = df[mask_complete].copy()
    missing = df[df[valeur_de_travail].isna()].copy()

    # SÃĐcuritÃĐ : Si pas assez de points pour le KNN
    if len(complete) < k:
        # On remplit avec la moyenne simple du fichier pour ÃĐviter le crash
        return df.fillna({valeur_de_travail: df[valeur_de_travail].mean()})

    # Normalisation
    scaler = StandardScaler()
    # .values pour ÃĐviter le warning "feature names"
    complete_scaled = scaler.fit_transform(complete[features].values)

    # Imputation
    for idx in missing.index:
        # On rÃĐcupÃĻre les features de la ligne Ã  remplir
        row_data = df.loc[[idx], features]
        
        # Si la ligne manque de features (ex: pas de lag), on passe ou on met la moyenne
        if row_data.isna().any().any():
            df.loc[idx, valeur_de_travail] = complete[valeur_de_travail].mean()
            continue
            
        row_scaled = scaler.transform(row_data.values)
        
        # Filtrage temporel si demandÃĐ
        if past_only:
            time_mask = complete["time"] < df.loc[idx, "time"]
            valid_subset = complete[time_mask]
            valid_scaled_subset = complete_scaled[time_mask.values]
            
            if len(valid_subset) < k: # Pas assez de passÃĐ ? On prend ce qu'on a
                if len(valid_subset) == 0: continue
                current_k = len(valid_subset)
            else:
                current_k = k
        else:
            valid_subset = complete
            valid_scaled_subset = complete_scaled
            current_k = k
        
        # Calcul de la distance Euclidienne : $d = \sqrt{\sum (x_i - y_i)^2}$
        distances = np.sqrt(((valid_scaled_subset - row_scaled)**2).sum(axis=1))
        k_idx = np.argsort(distances)[:current_k]
        
        # Moyenne des K plus proches voisins
        df.loc[idx, valeur_de_travail] = valid_subset.iloc[k_idx][valeur_de_travail].mean()
    
    return df[valeur_de_travail].to_numpy()
```

### Bootstrap saisonnier

La mÃĐthode dâimputation saisonniÃĻre par profil historique consiste Ã  remplir les valeurs manquantes dâune sÃĐrie temporelle en utilisant les comportements passÃĐs ou futur de la variable. Pour chaque jour de lâannÃĐe, on calcule la valeur typique, par exemple la mÃĐdiane sur toutes les annÃĐes disponibles, puis on remplace les donnÃĐes manquantes par cette valeur correspondant au mois considÃĐrÃĐ.

```python
def bootstrap_saisonnier_impute(df:pd.DataFrame, valeur_de_travail:str)->np.ndarray:
    """
    Impute les valeurs manquantes d'une colonne temporelle en utilisant un profil 
    saisonnier basÃĐ sur la mÃĐdiane historique par jour de l'annÃĐe.

    Args:
        df (pd.DataFrame): Dataset Ã  complÃĐter
        valeur_de_travail (str): Valeur que l'on veut interpolÃĐ

    Returns:
        np.ndarray: Le dataset complÃĐtÃĐ sur la valeur demandÃĐe
    """
        
    df = df.copy()
    df['time'] = pd.to_datetime(df['time'])
    
    # Extraire le jour et le mois pour dÃĐfinir la "saison"
    df['day_of_year'] = df['time'].dt.dayofyear
    
    # CrÃĐer le profil historique (la signature de la station)
    # On calcule la mÃĐdiane pour chaque jour de l'annÃĐe sur toutes les annÃĐes dispos
    profil_saisonnier = df.groupby('day_of_year')[valeur_de_travail].median()
    
    # SÃĐcuritÃĐ : si certains jours n'ont AUCUNE donnÃĐe historique, 
    # on lisse le profil avec une moyenne glissante
    profil_saisonnier = profil_saisonnier.interpolate(method='linear').fillna(df[valeur_de_travail].mean())
    
    # Imputation
    # On crÃĐe une Series qui contient la valeur du profil pour chaque ligne du DF original
    df['valeur_type'] = df['day_of_year'].map(profil_saisonnier)
    
    # On ne remplit que les vides
    df[valeur_de_travail] = df[valeur_de_travail].fillna(df['valeur_type'])
    
    return df[valeur_de_travail].to_numpy()
```

## Evaluation des mÃĐthodes

### MAE - Mean Absolute Error

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|
$$

OÃđ :

- $y_i$ : valeur observÃĐe  
- $\hat{y}_i$ : valeur prÃĐdite  
- $n$ : nombre d'observations  

```python
def nmae(prediction, reelle):
    range_val = np.max(reelle)
    if range_val == 0: return 0 # SÃĐcuritÃĐ si le signal est plat
    return np.mean(np.abs(reelle - prediction)) / range_val
```

### RMSE - Root Mean Square Error

$$
\text{RMSE} = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2 }
$$

OÃđ :

- $y_i$ : valeur observÃĐe  
- $\hat{y}_i$ : valeur prÃĐdite  
- $n$ : nombre d'observations  

```python
def nrmse(prediction, reelle):
    range_val = np.max(reelle)
    if range_val == 0: return 0
    # On calcule la RMSE standard puis on normalise
    return np.sqrt(np.mean((reelle - prediction) ** 2)) / range_val
```

### NSE - NashâSutcliffe Efficiency

$$
\text{NSE} = 1 - 
\frac{
\sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2
}{
\sum_{i=1}^{n} \left( y_i - \bar{y} \right)^2
}
$$

OÃđ :

- $y_i$ : valeur observÃĐe  
- $\hat{y}_i$ : valeur prÃĐdite  
- $\bar{y}$ : moyenne des valeurs observÃĐes  
- $n$ : nombre d'observations  

```python
def nse(prediction, reelle):    
    numerator = np.sum((reelle - prediction) ** 2)
    denominator = np.sum((reelle - np.mean(reelle)) ** 2)
    
    if denominator == 0:
        return np.nan
    
    return 1 - (numerator / denominator)
```

```python
def get_csv_files(folder_path):
    pattern = os.path.join(folder_path, "*.csv")
    files = glob.glob(pattern)

    if not files:
        print(f"â Aucun fichier CSV trouvÃĐ dans : {folder_path}")
        return []

    return files

def prepare_dataset(file, valeur_de_travail, remove_pct, rng, annee_deb = 0, annee_fin = 0):
    ds_name = os.path.basename(file)

    df = pd.read_csv(file, sep=';')

    if valeur_de_travail not in df.columns:
        print(f"â ïļ {ds_name} ignorÃĐ : colonne absente.")
        return None

    df['time'] = pd.to_datetime(df['time'])
    df = df.sort_values('time').reset_index(drop=True)

    y_full = df[valeur_de_travail].to_numpy()

    if np.isnan(y_full).all():
        print(f"â ïļ {ds_name} ignorÃĐ : uniquement des NaN.")
        return None

    n_points = len(df)

    y = y_full.copy()
    if annee_deb != annee_fin and annee_deb < annee_fin :
        mask = (df['time'].dt.year >= annee_deb) & (df['time'].dt.year <= annee_fin)
        y[mask] = np.nan


    if remove_pct > 0:
        n_remove = int(n_points * remove_pct)
        remove_idx = rng.choice(n_points, size=n_remove, replace=False)
        y[remove_idx] = np.nan

    if np.count_nonzero(~np.isnan(y)) < 4:
        print(f"â ïļ {ds_name} ignorÃĐ : pas assez de points valides.")
        return None

    return df, y_full, y, ds_name

def compute_interpolations(df, y, valeur_de_travail, models):
    """
    Calcule plusieurs interpolations/completion pour une colonne donnÃĐe
    et renvoie un dictionnaire {mÃĐthode: np.array}.
    """
    # On crÃĐe une copie avec les trous simulÃĐs
    df_with_holes = df.copy()
    df_with_holes[valeur_de_travail] = y
    df_with_holes['time_num'] = df['time'].astype('int64') // 10**9

    features = ["niveau_nappe_eau","lon","lat","time_num","ETP_Q","PRELIQ_Q","T_Q","surface_imp","surface_totale"]
    
    methods = {
        "Linear": interpolation_lineaire_array(df_with_holes, valeur_de_travail),
        "PCHIP": interpolation_pchip_array(df_with_holes, valeur_de_travail),
        "Akima": interpolation_akima_array(df_with_holes, valeur_de_travail),
        "Cubic": interpolation_cubique_array(df_with_holes, valeur_de_travail),
        "Poly": interpolation_polynomiale_array(df_with_holes, valeur_de_travail),
        "B-Spline": interpolation_spline_array(df_with_holes, valeur_de_travail),
        "KNN": knn_impute(df_with_holes, valeur_de_travail),
        "Bootstrap saisonier": bootstrap_saisonnier_impute(df_with_holes, valeur_de_travail),
        "Random forest": random_forest_delta_array(df_with_holes, valeur_de_travail),
        "LSTM" : lstm_predict_array(df_with_holes, models["LSTM"], mon_scaler, features, window_size=60, target_col=valeur_de_travail),
        "BILSTM" : lstm_predict_array(df_with_holes, models["BILSTM"], mon_scaler, features, window_size=6, target_col=valeur_de_travail),
        "LSTM2" : lstm_predict_array(df_with_holes, models["LSTM2"], mon_scaler, features, window_size=60, target_col=valeur_de_travail),
        "CNN" : lstm_predict_array(df_with_holes, models["CNN"], mon_scaler, features, window_size=6, target_col=valeur_de_travail),
    }


    return methods

def evaluate_methods(methods_dict, y_full, ds_name, remove_pct):
    rows = []
    
    y_full = y_full[6:]

    for method_name, arr in methods_dict.items():
        arr = arr[6:]
        rows.append({
            'dataset': ds_name,
            'method': method_name,
            'pct_removed': remove_pct,
            'NMAE': nmae(arr, y_full),
            'NRMSE': nrmse(arr, y_full)
        })

    return rows

def evaluate_all_files(folder_path, valeur_de_travail,
                       remove_pct_list=[0.1],
                       random_state=42,
                       max_files=1000):

    files = get_csv_files(folder_path)
    if not files:
        return None

    rng = np.random.default_rng(random_state)
    all_rows = []
    processed_count = 0

    models = {
        "LSTM" : load_model("./models/LSTM.keras", custom_objects={'masked_mse': masked_mse}),
        "BILSTM" : load_model("./models/BILSTM.keras", custom_objects={'masked_mse': masked_mse}),
        "LSTM2" : load_model("./models/LSTM2.keras", custom_objects={'masked_mse': masked_mse}),
        "CNN" : load_model("./models/CNN.keras", custom_objects={'masked_mse': masked_mse})
    }


    for remove_pct in remove_pct_list:
        for file in files:
            if processed_count >= max_files:
                break

            try:
                prepared = prepare_dataset(file, valeur_de_travail, remove_pct, rng)
                if prepared is None:
                    continue

                df, y_full, y, ds_name = prepared

                methods = compute_interpolations(df, y, valeur_de_travail, models)

                rows = evaluate_methods(methods, y_full, ds_name, remove_pct)
                all_rows.extend(rows)

                processed_count += 1

            except Exception as e:
                print(f"ðĨ Erreur sur {file} : {e}")
                continue

    if not all_rows:
        print("ð Aucun fichier traitÃĐ.")
        return None

    return pd.DataFrame(all_rows)

def plot_scatter_zooms(error_df, processed_count):
    methods_unique = error_df['method'].unique()
    palette = sns.color_palette("tab10", n_colors=len(methods_unique))
    markers = ['o', '^', 's', 'D', 'v', 'P', 'X', '*']

    zooms = [1, 0.05, 0.005]

    for z_limit in zooms:
        plt.figure(figsize=(10, 6))

        for i, method in enumerate(methods_unique):
            sub = error_df[error_df['method'] == method]
            plt.scatter(sub['NMAE'], sub['NRMSE'],
                        color=palette[i%len(markers)],
                        marker=markers[i%len(markers)],
                        s=100,
                        edgecolor='k',
                        alpha=0.6,
                        label=method)

        plt.xlim(0, z_limit)
        plt.ylim(0, z_limit)
        plt.xlabel("NMAE")
        plt.ylabel("NRMSE")
        plt.title(f"Zoom [0 - {z_limit}] ({processed_count} fichiers)")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

def compute_ranking_stats(error_df):
    error_df = error_df.copy()
    error_df['rank'] = error_df.groupby(['dataset', 'pct_removed'])['NMAE'].rank(method='min')

    stats = []

    methods = error_df['method'].unique()

    for method in methods:
        m_df = error_df[error_df['method'] == method]
        last_rank_per_group = error_df.groupby(['dataset', 'pct_removed'])['rank'].transform('max')
        stats.append({
            'method': method,
            'Premier': (m_df['rank'] == 1).sum(),
            'Dernier': (m_df['rank'] == last_rank_per_group[m_df.index]).sum(),
            'rank_moyen': m_df['rank'].mean()
        })

    return pd.DataFrame(stats)

def plot_first_vs_last(stats_df, processed_count):
    plt.figure(figsize=(10, 7))

    methods = stats_df['method'].unique()
    palette = sns.color_palette("tab10", n_colors=len(methods))
    markers = ['o', '^', 's', 'D', 'v', 'P', 'X', '*']

    for i, method in enumerate(methods):
        row = stats_df[stats_df['method'] == method]

        x_val = row['Premier'].iloc[0]
        y_val = row['Dernier'].iloc[0]

        plt.scatter(
            x_val,
            y_val,
            color=palette[i],
            marker=markers[i% len(markers)],
            s=250,
            edgecolor='k',
            label=method
        )

        plt.text(x_val + 0.1, y_val + 0.1,
                 method,
                 fontsize=10,
                 fontweight='bold')

    plt.xlabel("Nombre de fois arrivÃĐe 1er")
    plt.ylabel("Nombre de fois arrivÃĐe dernier")
    plt.title(f"FiabilitÃĐ relative des mÃĐthodes\n(Total datasets : {processed_count})")

    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_first_vs_avg_rank(stats_df, processed_count, val):
    plt.figure(figsize=(10, 7))

    methods = stats_df['method'].unique()
    palette = sns.color_palette("tab10", n_colors=len(methods))
    markers = ['o', '^', 's', 'D', 'v', 'P', 'X', '*']

    for i, method in enumerate(methods):
        row = stats_df[stats_df['method'] == method]

        x_val = row['Premier'].iloc[0]
        y_val = row['rank_moyen'].iloc[0]

        plt.scatter(
            x_val,
            y_val,
            color=palette[i],
            marker=markers[i % len(markers)],
            s=250,
            edgecolor='k',
            label=method
        )

        plt.text(x_val + 0.1, y_val + 0.02,
                 method,
                 fontsize=10,
                 fontweight='bold')

    plt.xlabel("Nombre de victoires (1er)")
    plt.ylabel("Rang moyen (plus petit = meilleur)")
    plt.title(f"Performance globale sur {val}\n({processed_count} fichiers)")

    plt.gca().invert_yaxis()
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_dataset_focus(error_df, target_dataset_name):
    df_graph = error_df[error_df['dataset'] == target_dataset_name]

    if df_graph.empty:
        print(f"â ïļ Dataset {target_dataset_name} non trouvÃĐ.")
        return

    plt.figure(figsize=(12, 8))

    methods = df_graph['method'].unique()
    pct_values = sorted(df_graph['pct_removed'].unique())

    palette = sns.color_palette("husl", n_colors=len(methods))
    markers = ['o', '^', 's', 'D', 'v', 'P', 'X', '*']

    marker_map = {
        pct: markers[i % len(markers)]
        for i, pct in enumerate(pct_values)
    }

    for i, method in enumerate(methods):
        for pct in pct_values:
            subset = df_graph[
                (df_graph['method'] == method) &
                (df_graph['pct_removed'] == pct)
            ]

            if not subset.empty:
                plt.scatter(
                    subset['NRMSE'],
                    subset['NMAE'],
                    color=palette[i],
                    marker=marker_map[pct],
                    s=150,
                    edgecolor='black',
                    linewidth=1,
                    alpha=0.9
                )

    # --- LÃGENDES ---
    from matplotlib.lines import Line2D

    legend_methods = [
        Line2D([0], [0],
               marker='o',
               color='w',
               label=m,
               markerfacecolor=palette[i],
               markersize=10)
        for i, m in enumerate(methods)
    ]

    legend_pct = [
        Line2D([0], [0],
               marker=marker_map[p],
               color='w',
               label=f"{int(p*100)}% trous",
               markerfacecolor='gray',
               markersize=10)
        for p in pct_values
    ]

    l1 = plt.legend(
        handles=legend_methods,
        title="MÃĐthodes",
        loc='upper left',
        bbox_to_anchor=(1, 1)
    )
    plt.gca().add_artist(l1)

    plt.legend(
        handles=legend_pct,
        title="% donnÃĐes retirÃĐes",
        loc='lower left',
        bbox_to_anchor=(1, 0)
    )

    plt.xlabel("NRMSE (sensibilitÃĐ aux grosses erreurs)")
    plt.ylabel("NMAE (erreur moyenne globale)")
    plt.title(f"Analyse de robustesse : {target_dataset_name}")

    plt.grid(True, linestyle='--', alpha=0.4)
    plt.tight_layout()
    plt.show()

def plot_error_heatmaps(error_df, error_metrics=['NMAE', 'NRMSE']):
    datasets = sorted(error_df['dataset'].unique())

    for metric in error_metrics:
        if metric not in error_df.columns:
            print(f"â ïļ MÃĐtrique {metric} absente dans le DataFrame.")
            continue

        # Pivot table : datasets en y, methods en x
        pivot_df = error_df.pivot_table(
            index='dataset',
            columns='method',
            values=metric,
            aggfunc='mean'  # moyenne si plusieurs pourcents retirÃĐs
        )

        plt.figure(figsize=(12, max(6, len(datasets)*0.3)))
        sns.heatmap(
            pivot_df,
            vmin=0,
            vmax=0.5,
            square=True,
            cbar_kws={'label': metric},
            linewidths=0
        )
        plt.title(f"Heatmap {metric} par dataset et mÃĐthode")
        plt.xlabel("MÃĐthode")
        plt.ylabel("Dataset")
        plt.tight_layout()
        plt.show()

def plot_mean_errors_by_method(error_df, processed_count, valeur_de_travail):
    """
    Graphique montrant l'erreur moyenne (NMAE et NRMSE)
    pour chaque mÃĐthode sur l'ensemble des datasets.
    """

    mean_df = (
        error_df
        .groupby('method')[['NMAE', 'NRMSE']]
        .mean()
        .reset_index()
    )

    plt.figure(figsize=(12,7), constrained_layout=True)

    methods = mean_df['method']
    palette = sns.color_palette("tab10", n_colors=len(methods))
    markers = ['o', '^', 's', 'D', 'v', 'P', 'X', '*']

    for i, row in mean_df.iterrows():
        plt.scatter(
            row['NMAE'],
            row['NRMSE'],
            color=palette[i],
            marker=markers[i % len(markers)],
            s=250,
            edgecolor='black'
        )

        plt.text(
            row['NMAE'],
            row['NRMSE'],
            row['method'],
            fontsize=11,
            fontweight='bold'
        )

    plt.xlabel("NMAE moyen")
    plt.ylabel("NRMSE moyen")
    plt.title(f"Erreur moyenne par mÃĐthode sur {valeur_de_travail}\n({processed_count} datasets)")

    plt.grid(True, linestyle='--', alpha=0.4)
    plt.show()

def run_full_evaluation(folder_path,
                        valeur_de_travail,
                        remove_pct_list=[0.1],
                        random_state=42,
                        target_dataset_name=None):
    
    print(f"""
============================================================================\n
\n
                            {valeur_de_travail}\n
\n
============================================================================\n
""")

    error_df = evaluate_all_files(
        folder_path,
        valeur_de_travail,
        remove_pct_list,
        random_state
    )

    if error_df is None:
        return None

    processed_count = error_df['dataset'].nunique()

    # Graph 1 & 2
    plot_scatter_zooms(error_df, processed_count)

    # Ranking
    stats_df = compute_ranking_stats(error_df)

    # Graph 3
    plot_first_vs_last(stats_df, processed_count)

    # Graph 4
    plot_first_vs_avg_rank(stats_df, processed_count, valeur_de_travail)

    plot_mean_errors_by_method(error_df, processed_count, valeur_de_travail)

    # Graph 5
    if target_dataset_name is None:
        target_dataset_name = error_df['dataset'].iloc[0]

    plot_error_heatmaps(error_df, error_metrics=['NMAE', 'NRMSE'])

    

    # plot_dataset_focus(error_df, target_dataset_name)

    return error_df, stats_df
```
